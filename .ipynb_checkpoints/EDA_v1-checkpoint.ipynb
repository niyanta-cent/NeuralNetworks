# 1. Introduction

### 1.1 About the problem
There is a need to develop a software product that can give the Toronto Police Department and the residents of the neighbourhood an idea about the likelihood of bicycle theft. This would  police department it would assist them to undertake stringent measures of anti-theft in certain neighborhoods. For the general public, it would help them be more precautious about their bicycle.

### 1.2 - The Dataset Source

The dataset was obtained from https://data.torontopolice.on.ca/datasets/bicycle-thefts. 
This dataset contains occurrences related to bicycle thefts from 2014 to 2020.

### 1.3 General Information
Original format: csv

**Dataset shape**: 25569 x 35 (rows x columns)


**Categorical Features**
- _Primary_Offence_: Primary Offence Type. There are 66 types of thefts.

## 2.1 Analyse Missing Values
Next, let us compute the percentage of Missing values and the percentage of Unique values for all features in the data.

```python
allcolumns = df.columns

unique_counts = []
missing_counts = []

for col in allcolumns:
    unique_counts.append(df[col].nunique())
    missing_counts.append(round((df[col].isna().sum() * 100)/df.shape[0],3))
    

data = {'Feature': allcolumns, '# of unique values': unique_counts, '% of Missing values': missing_counts}
df_counts  = pd.DataFrame.from_dict(data)

df_counts.sort_values(by='% of Missing values', ascending=False)
```

From the above output, following are our observations:
1.   **Bike_Model** feature has 37.7% of values missing. Followed by **Bike_Colour**,**Cost_of_Bike** and **Bike_Make** that have   8%, 7% and 0.5% respectively.
2. **ObjectId2**, **OBJECTID**, **event_unique_id** have too many unique values(more than 89%). Therefore these features can be dropped.

# 3. Data Preparation and Cleaning

Following are our data cleaning techniques - 
1. Drop ObjectId2, OBJECTID, event_unique_id features as these have too many unique values.
2. Hood_ID which denotes 'NeighborhoodID where a homicide occurred' will be dropped as we already have the corresponding NeighbourhoodName.
3. We will drop attributes X and Y as these are the coordinates version of Lat/Long.
4. We will drop Report_Date and Occurence Data because their Year, Month, Day and Hour parts are already present as features.
5. Let us drop Report_Year, Occurrence_Year as these attributes are not related to the theft of a bike.
2. Impute Missing values in Bike_Model, Bike_Colour with most frequent values. 
3. Impute Missing values in Cost_of_Bike with it's median.
